```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import Table, Column, MetaData, Integer, String, Date, Float, BigInteger

# Part 1: Connect DB and create table
db_engine = create_engine("postgresql+psycopg2://postgres:admin1234@localhost:5432/bootcamp_2512")

metadata = MetaData()

persons_schema = Table(
  "persons",
  metadata,
  Column("id", BigInteger, primary_key=True),
  Column("name", String(20), nullable=False),
  Column("dob", Date, nullable=False),
  Column("salary", Float, nullable=False)
)

metadata.create_all(db_engine)

# Part 2: load data from csv

df_persons = pd.read_csv("persons.csv")

# ETL (Extract Transform Load)
df_persons['dob'] = pd.to_datetime(df_persons['dob'].astype(str), format="%Y-%m-%d").dt.date

# write to db
df_persons.to_sql("persons", db_engine, index=False, if_exists="replace")

# Read DB by Pandas
df_sql_result = pd.read_sql("select * from persons where salary > 20000", con=db_engine)
print(df_sql_result)
```

```python
import pandas as pd

# Series
series = pd.Series([2, 8, -2, 10])
print(type(series))
print(series)

staff = {
    "name": "Leo",
    "age": 23
}

df = pd.DataFrame(staff, index=[0])
print(df)

team = {
  "names" : ["Leo", "Tom", "Peter"],
  "ages" : [23, 10, 18]
}

df_team = pd.DataFrame(team)
print(df_team)

print(df_team['names'])

print(type(df_team['names']))

df_team['gender'] = pd.Series(['M', 'M', 'M'])
print(df_team)
```

```python
# loc -> dataframe
df_team_2 = df_team.loc[df_team["ages"] > 15]
print(df_team_2)
print(type(df_team_2)) # DataFrame

df_team_3 = df_team.loc[df_team["names"].str.len() > 3]
print(df_team_3)

# & (and), | (or)
df_team_4 = df_team.loc[(df_team["ages"] > 15) | (df_team["names"].str.len() > 3)]
print(df_team_4)

df_team_5 = df_team.loc[df_team["names"].str.contains("o")]
print(df_team_5)
```

```python
print(df_team.info())
print(df_team.describe())
print(df_team.head())
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)
print(type(df))
# print(df.head())

# Sex = male and Name contains "Allen"
print(df[(df['Sex'] == "female") & (df['Name'].str.contains("Allen"))])

print(df['Age'].isnull().any())

# Drop Data with null age
df_drop_na_age = df.dropna(subset=["Age"])
print(len(df_drop_na_age))
print(len(df))
print(df_drop_na_age['Age'].isnull().any())
```

```python
# Refill
df_for_refill = df.copy()
df_for_refill['Age'] = df_for_refill['Age'].fillna("N/A")
print(len(df_for_refill))
print(df_for_refill['Age'].isnull().any())
```

```python
# loc -> where
# group by

print(df['Age'].mean())
print(df['Age'].min())
print(df['Age'].max())

df_result = df.groupby('Survived')['Age'].mean()
print(df_result)
print(type(df_result))
```

```python
from datetime import datetime, date, timedelta

print(datetime.now())
print(date.today())

birthday = date(2020, 12, 31)
print(birthday)

birthday = birthday + timedelta(days=10)
print(birthday)
```

```python
from dateutil.relativedelta import relativedelta

probation_end_date = date.today() + relativedelta(months=3)
print(probation_end_date)

print(date.today() + relativedelta(years=2))
print(date.today() + relativedelta(days=90))
```

```python

formatted_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
print(formatted_date)
print(type(formatted_date))

str_datetime = "2025/12/31 23:59:59"
today = datetime.strptime(str_datetime, "%Y/%m/%d %H:%M:%S")
print(today)
print(type(today))
```

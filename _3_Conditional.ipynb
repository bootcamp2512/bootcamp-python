```python
score = 91
# >= 90 (A)
# >= 80 (B)
# (F)
# Grade=A
if (score >= 90):
  grade = 'A'
elif (score >= 80):
  grade = 'B'
else:
  grade = 'F'
print(f"Grade={grade}")
```

```python
x = 10

# Java: || &&
# Pythob: or and

if (x > 4 and x < 12):
  print('hello')

if (x > 4 or x < 9):
  print('goodbye')

age = 66

# Java: age >= 65 ? 2 : 10
fee = 2 if age >= 65 else 10
print(fee)
```

```python
# Java: equals
# Python: ==
input_password = 'admin1234'
actual_password = 'admin1234'
if (input_password == actual_password):
  print('Login Success.')
else:
  print('Login Fail.')

print(input_password != actual_password) # False
```

```python
# Number Math Operaion
# odd
number = 3
if (number % 2 == 1):
  print("Odd.")
else:
  print("Even.")
```

```python
# Python >= 3.10
# Java switch
# Python match
weekday = 'Thursday'
match weekday:
  case 'Monday':
    print('Mon')
  case 'Tuesday':
    print('Tue')
  case 'Wednesday':
    print('Wed')
  case 'Thursday':
    print('Thu')
  case 'Friday':
    print('Fri')
  case _:
    print('Default')
```

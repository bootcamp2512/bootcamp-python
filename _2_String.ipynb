```python
# array, String
word = 'Python'

print(word[0])
print(word[-1]) # Java: length - 1
print(word[-2])

print(word[:3])
print(word[0:3])

print(word[3:])
print(word[2:3])

# str
print(word.upper())
print(word.lower())
print(word.replace('y', 'a'))
```

```python
# str method
student_name = 'Steven'
print(student_name.count('e')) # 2
print(student_name.isalpha()) # True
print(student_name.isdigit()) # False
print(student_name.startswith('s')) # False
print(student_name.startswith('S')) # True, case sensitive

score = "93"
print(score.isdigit()) # True
```
